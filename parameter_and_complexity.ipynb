```python
import matplotlib.pyplot as plt
import random

x = [x for x in range(30)]
y = []

for i in x:
  y.append(10*i+10*i*i+10*i*i*i) #10 10 10
plt.plot(x,y,'r-') #red

y4 = []
for i in x:
  y4.append(1*i+1*i*i+10*i*i*i) #1 1 10
plt.plot(x,y4,'yo') #yellow point

y5 = []
for i in x:
  y5.append(1*i+1*i*i+5*i*i*i) #1 1 5
plt.plot(x,y5,'bo') #blue
#最高阶的参数影响复杂度,低阶的参数影响很小，可忽略（红黄重合）
#high comlexity => more curvly

y2 = []
for i in x:
  y2.append(5*i+5*i*i+5*i*i*i) #5 5 5
plt.plot(x,y2,'g-') #green

y3 = []
for i in x:
  y3.append(1*i+1*i*i+1*i*i*i) #1 1 1
plt.plot(x,y3,'b-') #blue
plt.show()
```
